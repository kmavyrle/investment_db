```python
import yfinance as yf
import pandas as pd

def get_yahoo_prices(tickers, start_date, end_date=None, field="Adj Close"):
    """
    tickers: list of tickers, e.g. ["AAPL", "MSFT", "^GSPC"]
    start_date: "YYYY-MM-DD"
    end_date: optional, default = today
    field: "Adj Close", "Close", "Open", "High", "Low", "Volume"
    """
    data = yf.download(
        tickers=tickers,
        start=start_date,
        end=end_date,
        auto_adjust=False,
        progress=False
    )

    # Single ticker returns a Series → force DataFrame
    if isinstance(data, pd.Series):
        data = data.to_frame()

    # Multi-ticker case
    if isinstance(data.columns, pd.MultiIndex):
        prices = data[field]
    else:
        prices = data[[field]]

    prices.index = pd.to_datetime(prices.index)
    prices = prices.dropna(how="all")

    return prices



```

```python
tickers = ['XPEV',"UEC","WLN.PA","SHLD","SBET","PDD","LKNCY","JD","GRAB","INDY","DSCSY","DIDIY","CHA","CEG","BILI","BABA","BIDU",
           "9880.HK","9698.HK","81810.HK","6862.HK","600875.SS","3690.HK","0020.HK","1164.HK","002594.SZ","000333.SZ","SPY"]
df = get_yahoo_prices(tickers,"2017-01-01")
```
