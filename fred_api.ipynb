```python
def get_fred_data(ticker,name):
    data = fred.get_series_all_releases(ticker)#.tail(4*10)
    data = data.groupby('date').first()
    #print(data)
    data = data[['realtime_start','value']]
    data.set_index('realtime_start',inplace = True)
    #data = data.tail(4*10)
    #data = data.pct_change().fillna(0)*100
    data = data[~data.index.duplicated(keep='last')]
    data['Name'] = [name]*data.shape[0]
    return data
```

```python
df = get_fred_data("CPIAUCSL",ticker_map['CPIAUSCL'])
df
```
